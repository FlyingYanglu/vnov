```python
import json
```

```python
with open("./datasets/yujie/characters_info.json", "r", encoding="utf-8") as f:
    char_dict = json.load(f)
char_dict
```

```python
def extract_appearance(appearance_dict):
    tags = []
    for key, value in appearance_dict.items():
        if key == 'Eyes':
            tags.append(value['Color'][0])
        elif isinstance(value, dict):
            # If the value is a nested dictionary, recursively extract tags
            tags.extend(extract_appearance(value))
        elif isinstance(value, list):
            # If the value is a list, extend the tags list with it
            tags.extend(value)
        else:
            # Otherwise, just add the value as a tag
            tags.append(value)
    return tags

# Convert the characters' appearance data
def convert_to_tags_format(characters_dict):
    result = {}
    for character, details in characters_dict.items():
        appearance = details.get('Appearance', {})
        tags = ['1full body image']
        if details['Gender'] == "Male":
            tags.append('1boy')
        else:
            tags.append('1girl')

        tags += extract_appearance(appearance)
        
        result[character] = {'prompt': (', '.join(tags) + ", anime, cute, japanese manga style "), "num_occurence": details["num_occurence"]}
    return result
# appearance_combined[character] = {'prompt': ("full body image, " + ', '.join(combined_appearance) + ", anime, cute, japanese manga style --no nsfw").replace("petite", "small"), "num_occurence": details["num_occurence"]}
    
```

```python
mj_char_info = convert_to_tags_format(char_dict)
mj_char_info
```

```python
with open("./datasets/yujie/midjourney_char_info.json", "w", encoding="utf-8") as f:
    json.dump(mj_char_info, f, ensure_ascii=False, indent=4)
```
